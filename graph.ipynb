```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("share-of-population-who-have-ever-been-homeless-2003.csv")
```

```python
# plt.barh(df['Entity'], df['Share of population with an episode of homelessness (Toro et al. (2007))'])

fig, ax = plt.subplots()
ax.barh(df['Entity'], df['Share of population with an episode of homelessness (Toro et al. (2007))'])
ax.set_title('Share of population who have ever been homeless, 2003')

# add labels on each bar with the number
for i, v in enumerate(df['Share of population with an episode of homelessness (Toro et al. (2007))']):
    ax.text(v + 0.1, i, str(v) + '%')
```
